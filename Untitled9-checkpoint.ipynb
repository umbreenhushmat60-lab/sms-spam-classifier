# Step 4: Text Preprocessing

```python
# Import required libraries for text processing
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
# Initialize Porter Stemmer
ps = PorterStemmer()
```

```python
# Define text transformation function
def transform_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Tokenization
    text = nltk.word_tokenize(text)
    
    # Remove non-alphanumeric characters
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    # Remove stopwords and punctuation
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    
    # Stemming
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
# Test the transformation function
transform_text("Did you like my presentation in Machine Learning")
```

```python
# Apply transformation to entire dataset
df['transformed_text'] = df['text'].apply(transform_text)
```

# Step 6: Feature Extraction and Model Building


```python
# Create corpus for spam messages
spam_corpus = []
for msg in df[df['target'] == 1]['transformed_text'].tolist():
    for word in msg.split():
        spam_corpus.append(word)
```

```python
# Check length of spam corpus
len(spam_corpus)
```

```python
# Plot most common words in spam
from collections import Counter
spam_common = Counter(spam_corpus).most_common(30)
df_spam = pd.DataFrame(spam_common, columns=['word', 'count'])
```

```python
# Create corpus for ham messages
ham_corpus = []
for msg in df[df['target'] == 0]['transformed_text'].tolist():
    for word in msg.split():
        ham_corpus.append(word)
```

```python
# Check length of ham corpus
len(ham_corpus)
```

```python
# Plot most common words in ham
ham_common = Counter(ham_corpus).most_common(30)
df_ham = pd.DataFrame(ham_common, columns=['word', 'count'])

plt.figure(figsize=(12, 8))
sns.barplot(x='word', y='count', data=df_ham)
plt.xticks(rotation=90)
plt.title('Top 30 Words in Ham Messages')
plt.tight_layout()
plt.show()
```

```python
# Import feature extraction tools
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# Initialize TfidfVectorizer
tfidf = TfidfVectorizer(max_features=3000)
```

```python
# Transform text data to features
X = tfidf.fit_transform(df['transformed_text']).toarray()
```

```python
# Check feature matrix shape
X.shape
```

```python
# Create target variable
y = df['target'].values
```

```python
# View target array
y
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# Import classifiers and metrics
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
# Initialize classifiers
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

# Step 7: Model Evaluation


```python
# Cell 71: Train and evaluate Gaussian Naive Bayes
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)

print("Gaussian Naive Bayes Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_gnb):.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_gnb))
print(f"Precision: {precision_score(y_test, y_pred_gnb):.4f}")
print("-" * 50)
```

```python
# Cell 72: Train and evaluate Multinomial Naive Bayes
mnb.fit(X_train, y_train)
y_pred_mnb = mnb.predict(X_test)

print("Multinomial Naive Bayes Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_mnb):.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_mnb))
print(f"Precision: {precision_score(y_test, y_pred_mnb):.4f}")
print("-" * 50)
```

```python
# Cell 73: Train and evaluate Bernoulli Naive Bayes
bnb.fit(X_train, y_train)
y_pred_bnb = bnb.predict(X_test)

print("Bernoulli Naive Bayes Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_bnb):.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_bnb))
print(f"Precision: {precision_score(y_test, y_pred_bnb):.4f}")
print("-" * 50)
```
